```python
X = diabetes_dataset.drop(columns=['Outcome'],axis=1)
y = diabetes_dataset['Outcome']
```

```python
print(accuracy_score(Y_test,y_pred))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

X_test = np.random.rand(100, 2)  # Example test data
Y_test = np.random.randint(0, 2, 100)  # Example binary labels

acc = pd.Series(dtype=float)  # Initialize an empty Series
X_axis = list(range(1, 31))

for i in X_axis:
    knn_model = KNeighborsClassifier(n_neighbors=i)
    knn_model.fit(X_test, Y_test)
    prediction = knn_model.predict(X_test)
    acc = pd.concat([acc, pd.Series(metrics.accuracy_score(Y_test, prediction))], ignore_index=True)

plt.plot(X_axis, acc)
plt.xticks(X_axis)
plt.title("Finding best value for k (n_neighbors)")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.show()
```

```python
print('Highest value: ',acc.values.max())
```
